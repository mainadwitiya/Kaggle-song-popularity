```python
import sys
sys.path.append("kuma_utils/")
from kuma_utils.preprocessing.imputer import LGBMImputer


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_base = pd.read_csv('https://raw.githubusercontent.com/mainadwitiya/Kaggle-song-popularity/main/csv/train.csv?token=GHSAT0AAAAAABPK7AEBVULXIUG5MRGDGVFGYPWRUNQ')
test_base = pd.read_csv('https://raw.githubusercontent.com/mainadwitiya/Kaggle-song-popularity/main/csv/test.csv?token=GHSAT0AAAAAABPK7AEA6PG33UFDPGL462I4YPWRVWQ')
submission = pd.read_csv('https://raw.githubusercontent.com/mainadwitiya/Kaggle-song-popularity/main/csv/sample_submission.csv?token=GHSAT0AAAAAABPK7AEATANZGA77FJZG6ZX6YPWZEGA')
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
cols = ['acousticness', 'liveness', 'loudness']
id_and_pop = ['id', 'song_popularity']
```

```python
pt = PowerTransformer()
train_base[cols] = pt.fit_transform(train_base[cols])
test_base[cols] = pt.transform(test_base[cols])
```

```python
lgbm_cols = [col for col in train_base.columns if (col not in cols) and (col not in id_and_pop)]
```

```python
lgbm_cols
```

```python
lgbm_imtr = LGBMImputer(n_iter=400, verbose=True)

train_lgbmimp = lgbm_imtr.fit_transform(train_base[lgbm_cols])
test_lgbmimp = lgbm_imtr.transform(test_base[lgbm_cols])
```

```python
for col in train_base[cols]:
  train_lgbmimp[col] = train_base[col] 
  test_lgbmimp[col] = test_base[col]
```

```python
for col in train_lgbmimp.columns:
  sns.distplot(train_lgbmimp[col])
  plt.show()
```

```python
train_lgbmimp.describe()
```

```python
log1p_cols = ['energy', 'instrumentalness']
log_cols = ['song_duration_ms', 'danceability', 'speechiness', 'tempo', 'audio_valence']
```

```python
for col in log1p_cols:
  train_lgbmimp[col] = np.log1p(train_lgbmimp[col])
  test_lgbmimp[col] = np.log1p(test_lgbmimp[col])
```

```python
for col in log_cols:
  train_lgbmimp[col] = np.log(train_lgbmimp[col])
  test_lgbmimp[col] = np.log(test_lgbmimp[col])
```

```python
for col in train_lgbmimp.columns:
  sns.distplot(train_lgbmimp[col])
  plt.show()
```

```python
train_lgbmimp.to_csv('29jan_imputed_train.csv', index=None)
test_lgbmimp.to_csv('29jan_imputed_test.csv', index=None)
```
