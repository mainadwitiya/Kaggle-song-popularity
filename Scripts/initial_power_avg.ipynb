```python
import pandas as pd
import numpy as np
```

```python
#ordered by best scores
first_lgbm = pd.read_csv('lgbm-mmmm.csv')
second_xgb = pd.read_csv('XGB_21_jan.csv')
third_lgbm = pd.read_csv('submission_lgb.csv')
forth_lgbm = pd.read_csv('lgbm_2000_n_est.csv')
fifth_lgbm = pd.read_csv('lgbm_2000_n_est (1).csv')
sixth_xgb = pd.read_csv('submission_xgb_best_params1_250_trials_imputed_median.csv')
seventh_cat = pd.read_csv('submission_cat (1).csv')
eight_xgb = pd.read_csv('submission_xgb_best_params1_200_trials.csv')
ninth_gmy = pd.read_csv('Auto_gmy_submission.csv')
tenth_auto = pd.read_csv('AlteredAutoKaggle.csv')
eleventh_xgb = pd.read_csv('submission_xgb_best_params1_250_trials.csv')
twelveth_lgbm = pd.read_csv('submission_lgbm_best_params1_200_trials_imputed_median.csv')
thirtheen_lgbm = pd.read_csv('submission_lgbm_best_params1_200_trials_imputed_median_300_rounds.csv')
fourteen_pycaret = pd.read_csv('lgbmpycaret_submission(2).csv')
fifteen_ = pd.read_csv('sub.csv')
sixteen_ = pd.read_csv('cat_boost_tuned (1).csv')
```

```python
list_of_subs = [first_lgbm.song_popularity, 
                second_xgb.song_popularity, 
                third_lgbm.song_popularity, 
                forth_lgbm.song_popularity, 
                fifth_lgbm.song_popularity, 
                sixth_xgb.song_popularity,
                seventh_cat.song_popularity, 
                eight_xgb.song_popularity, 
                ninth_gmy.song_popularity, 
                tenth_auto.song_popularity,
                eleventh_xgb.song_popularity, 
                twelveth_lgbm.song_popularity,
                thirtheen_lgbm.song_popularity, 
                fourteen_pycaret.song_popularity, 
                fifteen_.song_popularity, 
                sixteen_.song_popularity,]

group_labels = [str(i) for i in range(1,17)]
```

```python
## correlations for best model

import seaborn as sns
import matplotlib.pyplot as plt

data = np.corrcoef(list_of_subs)

fig, ax = plt.subplots(figsize=(20,20))
im = ax.imshow(data)

ax.set_xticks(np.arange(len(group_labels)))
ax.set_yticks(np.arange(len(group_labels)))

ax.set_xticklabels(group_labels)
ax.set_yticklabels(group_labels)

for i in range(len(group_labels)):
    for j in range(len(group_labels)):
        text = ax.text(j, i, data[i, j].round(5),
                       ha="center", va="center")
```
